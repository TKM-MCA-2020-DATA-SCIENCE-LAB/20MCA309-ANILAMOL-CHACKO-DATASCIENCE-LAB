```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x ** 2
y
```

```python
plt.plot(x,y,'b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,'g--')
plt.subplot(1,3,2)
plt.plot(x,y,'b*-')
plt.subplot(1,3,3)
plt.plot(x,y,'r--')
plt.show()
```

```python
#using object oriented
fig=plt.figure()
axes=fig.add_axes([0,0.9,0.6,0.7])
axes.plot(x,y,'g')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.3,0.6,0.9,0.6])
axes2=fig.add_axes([0.4,0.8,0.2,0.3])
axes1.plot(x,y,'g')
axes1.set_xlabel('X Axis')
axes1.set_ylabel('Y Axis')
axes1.set_title('Title')
axes2.plot(x,y,'r')
axes2.set_xlabel('X Axis')
axes2.set_ylabel('Y Axis')
axes2.set_title('Title')
plt.show()
```

```python
#subplots
fig, axes= plt.subplots(1,5)
for axe in axes:
  axe.plot(x,y,'r--')
  axe.set_xlabel('X')
  axe.set_ylabel('Y')
  axe.set_title('title')
fig.tight_layout()
```

```python
fig, axes= plt.subplots(1,4)
```

```python
fig=plt.figure(figsize=(2,6),dpi=50)
plt.plot(x,y,'b')
plt.show()
```

```python
fig,axes=plt.subplots(figsize=(5,6))
axes.plot(x,y,'g')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Title')
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.3,0.6,0.9,0.6])
axes.plot(x,x**2,label='x**2')
axes.plot(x,x**3,label='x**3')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('title')
axes.legend(loc=2)
plt.show()
```
