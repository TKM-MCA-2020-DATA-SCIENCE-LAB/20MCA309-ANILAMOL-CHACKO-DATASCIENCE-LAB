```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x ** 2
y
```

```python
fig, axes= plt.subplots()
axes.plot(x, x**2, 'b.-')
axes.plot(x, x**3, 'g--')
```

```python
fig, axes= plt.subplots(figsize=(12,6))
axes.plot(x, x+1, color="blue", linewidth=0.25)
axes.plot(x, x+2, color="blue", linewidth=0.50)
axes.plot(x, x+3, color="blue", linewidth=1.00)
axes.plot(x, x+4, color="blue", linewidth=2.00)

axes.plot(x, x+5, color="red", lw=3, linestyle='-')
axes.plot(x, x+6, color="red", lw=3, ls='-.')
axes.plot(x, x+7, color="red", lw=3, ls=':')

line, =axes.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5,10,15,20])

axes.plot(x, x+9, color="green", lw=3, ls='-', marker='+')
axes.plot(x, x+10, color="green", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="green", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="green", lw=3, ls='--', marker='1')

axes.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="red")





```

DISTRIBUTION PLOTS
 1)distplot
 2)jointplot
 3)pairplot
 4)rugplot
 5)kdeplot

```python
import seaborn as sns
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

DISTPLOT

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['total_bill'], kde=False, color='green', bins=30)
```

JOINTPLOT

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='resid')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

PAIRPLOT

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue="sex", palette='coolwarm')
```

RUGPLOT

```python
sns.rugplot(tips['tip'])
```

```python
sns.rugplot(tips['total_bill'])
```
