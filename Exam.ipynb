K-Means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/College_Data')
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
a=df.iloc[:,2:]
a
```

```python
k=KMeans(n_clusters=2)
k.fit(a)
```

```python
k.cluster_centers_
```

```python
def Convertor(cluster):
  if(cluster=='yes'):
    return 1
  else:
    return 0
```

```python
df['Cluster']=df['Private'].apply(Convertor)
df.head()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(df['Cluster'],k.labels_))
```
